### 2.Feature Seclection

```python
X = df.drop('diagnosis',axis=1)
Y=df.diagnosis
```

```python
from sklearn.ensemble import ExtraTreesClassifier 
```

```python
model = ExtraTreesClassifier() 
model.fit(X, Y) 
```

```python
print(model.feature_importances_)
```

```python
fs_importance = pd.Series(model.feature_importances_,index = X.columns)
fs_importance.nlargest(10).plot(kind = 'barh')
plt.show()
```

### 3 Algorithm for classification

```python
X_Selected=df.loc[:, ['BB','GJ','HA','LE','KK','LG','HT','DZ','GL','GL','JN','KH','T','heart_rate','HR','EA','diagnosis']]
```

```python
X_Selected
```

```python
X_Selected.to_csv('SelectedDataset.csv')
```

```python
X=X_Selected.drop(['diagnosis'], axis=1)
Y=X_Selected[['diagnosis']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=365)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Random Forest classification

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
```

```python
RF.fit(X_train, Y_train)
```

```python
RF.score(X_test,Y_test)
```
